```python
from pprint import pprint

with open("show_vlan.txt") as f:
    show_vlan = f.read()

vlan_list = []

for line in show_vlan.splitlines():
    if 'VLAN' in line or '-----' in line or 'Et6' in line:
        continue
    fields = line.split()
    vlan_id = fields[0]
    vlan_name = fields[1]
    vlan_list.append((vlan_id,vlan_name))

pprint(vlan_list)
   
```

```python
with open("show_arp.txt") as m:
    show_arp = m.read()

found1, found2 = (False, False)
for line in show_arp.splitlines():
    
    if "protocol" in line.lower():
        continue
    
    fields = line.split()
    ip_add = fields[1]
    mac_add = fields[3]
    
    if ip_add == '10.220.88.1':
        print(f'Default gateway IP/Mac: {ip_add} / {mac_add}')
        found1 = True

    elif ip_add == '10.220.88.30':
        print(f'Arista3 IP/Mac is {ip_add} / {mac_add}')
        found2 = True
    
    if found1 and found2:
        print("Exiting...")
        break
```

```python
with open("show_lldp_neighbors_detail.txt") as n:
    show_lldp = n.read()

system_name, port_id = (None, None)

for line in show_lldp.splitlines():
    if "System Name" in line:
        _, system_name = line.split('System Name: ')
     
    elif "Port id" in line:
        _, port = line.split('Port id: ')
    
    if port and system_name:
        break

print()
print("System Name: {}".format(system_name))
print("Port ID: {}".format(port))       
```

```python
arp_table = [('10.220.88.1', '0062.ec29.70fe'),
 ('10.220.88.20', 'c89c.1dea.0eb6'),
 ('10.220.88.21', '1c6a.7aaf.576c'),
 ('10.220.88.28', '5254.aba8.9aea'),
 ('10.220.88.29', '5254.abbe.5b7b'),
 ('10.220.88.30', '5254.ab71.e119'),
 ('10.220.88.32', '5254.abc7.26aa'),
 ('10.220.88.33', '5254.ab3a.8d26'),
 ('10.220.88.35', '5254.abfb.af12'),
 ('10.220.88.37', '0001.00ff.0001'),
 ('10.220.88.38', '0002.00ff.0001'),
 ('10.220.88.39', '6464.9be8.08c8'),
 ('10.220.88.40', '001c.c4bf.826a'),
 ('10.220.88.41', '001b.7873.5634')]


for ip_addr, mac_addr in arp_table:
    mac_addr = mac_addr.split('.')
    mac_addr = "".join(mac_addr)
    mac_addr = mac_addr.upper()

    new_mac = []
    while len(mac_addr) > 0:
        part = mac_addr[:2]
        mac_addr = mac_addr[2:]
        new_mac.append(part)
        
    new_mac = ":".join(new_mac)
    print(new_mac)
```

```python
import os

base_ip = "10.10.100."
base_cmd = "ping -c 2"

ip_list = []

for last_octet in range(1, 255):
    new_ip = base_ip + str(last_octet)
    ip_list.append(new_ip)

for i, ip_addr in enumerate(ip_list):
    print(f'{i} ---> {ip_addr}')

ip_list = ip_list[2:6]

print("-" * 80)
for ip_addr in ip_list:
    print("IP Addr: ", ip_addr)
    return_code = os.system("{} {}".format(base_cmd, ip_addr))
    print("-" * 80)
print("-" * 80)
```
